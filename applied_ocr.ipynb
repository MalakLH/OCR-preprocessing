```python
import cv2
from PIL import Image
import pytesseract
```

```python
img_path= '/media/O.jpg'
image= Image.open(img_path)
```

```python
ocr_result= pytesseract.image_to_string(image)
print(ocr_result)
```

```python
#trying to separate the main text from the annotations

img= cv2.imread(img_path)

def grayscale(image):
  return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

img_gray= grayscale(img)

img_blurred= cv2.GaussianBlur(img_gray, (5,5), 0)

thresh = cv2.threshold(img_blurred, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1,10))

dilate = cv2.dilate(thresh, kernel, iterations=1)

cnts= cv2.findContours(dilate,cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

cnts= cnts[0] if len(cnts) == 2 else cnts[1]

cnts= sorted(cnts, key=lambda x: cv2.boundingRect(x)[0])
```

```python
#to visulize our contours we do:

image=cv2.imread(img_path)
for c in cnts:
  x, y, w, h = cv2.boundingRect(c)
  if h > 300 and w > 100:
    roi= image[y:y+h, x:x+w]
    cv2.imwrite('/media/image_roi.jpg',roi)
    cv2.rectangle(image, (x,y),(x+w,y+h),(36,255,12),2)

cv2.imwrite('/media/image_contours.jpg',image)
```

```python
ocr_better_result= pytesseract.image_to_string(Image.open('/media/image_roi.jpg'))
print(ocr_better_result)
```
